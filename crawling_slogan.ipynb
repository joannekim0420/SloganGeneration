```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from pyvirtualdisplay import Display

# display = Display(visible=0, size=(1024,768))
# display.start()
# selenium에서 사용할 웹 드라이버 절대 경로 정보

chrome_options = Options()  
chrome_options.add_argument("--headless")  
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument("--start-fullscreen")
chrome_options.add_argument("--no-proxy-server")
chrome_options.add_argument("--proxy-server='direct://'");
chrome_options.add_argument("--proxy-bypass-list=*");
chrome_options.add_experimental_option("detach",True)
chrome_options.add_experimental_option("excludeSwitches",["enable-logging"])

# browser = webdriver.Chrome(executable_path=chromedriver,options = chrome_options)
driver = webdriver.Chrome(service =Service(ChromeDriverManager().install()), options=chrome_options)
```

```python
categories = ["education-slogans", "food-slogans","restaurant-slogans","car-slogan","apparel-slogans",
              "technology-slogans","business-slogans","company-slogans","household-slogans","tours-slogans",
              "airlines-slogans","television-channels-slogan","financial-slogans","cosmetics-slogans","uncategorized",
             "drinking-slogans"]
```

```python
def get_companies(driver):
    div = driver.find_element("css selector", "div[class='col-sm-7 col-lg-5']")

    buttons = div.find_elements("css selector", "a.list-group-item")
    companies_url = [b.get_attribute('href') for b in buttons]
    return companies_url

def get_slogan(driver, url):
    driver.get(url)
    company_name = driver.find_element("css selector", "div.detail h1").text.replace('slogans','')
    slogans = driver.find_element("css selector", "div.info_content").text.split('\n')
    try:
        desc = driver.find_element("css selector", "div.text-info>p").text
    except:
        desc = ""
    return company_name, slogans, desc
```

```python
# selenum의 webdriver에 앞서 설치한 chromedirver를 연동한다.
driver = webdriver.Chrome(service =Service(ChromeDriverManager().install()), options=chrome_options)
driver.implicitly_wait(10)
f = open("./company_slogan_desc_crawl_drinking","w",encoding="utf-8")

# for category in categories[]:
#     find the number of pages
category = categories[-1]
category_url = 'https://www.sloganlist.com/'+category
driver.get(category_url)

last_page = int(driver.find_element("css selector", "div[class='list_page page_group']").text.split('\n')[0])//20 + 1 
for page in range(1, last_page+1):
    category_url = 'https://www.sloganlist.com/'+category+'/index_'+str(page)+".html" if page>1 else 'https://www.sloganlist.com/'+category

    driver.get(category_url)
    companies_url = get_companies(driver)

    for comp_url in companies_url:
        print(comp_url)
        company_name, slogans, desc = get_slogan(driver, comp_url)
        for slogan in slogans:
            f.write(company_name+"\t"+desc+"\t"+slogan+"\n")

f.close()
driver.quit()
```
